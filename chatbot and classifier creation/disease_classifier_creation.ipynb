```python
def remove_space_between_word(dataset):
    for col in dataset.columns:
        for i in range(len(dataset[col])):
            if (type(dataset[col][i]) == str ):
                dataset[col][i] = dataset[col][i].strip()
                dataset[col][i] = dataset[col][i].replace(" ", "_")
    return dataset
```

```python
def symptom_encoder(ds):
    for i in ds.columns:
        if i=='Disease':
            continue
        for j in range(len(ds[i])):
            try:
                ds[i][j]=symptom_sevirity[ds[i][j]]
            except Exception:
                pass
    ds=ds.fillna(0)        
    ds = ds.replace("foul_smell_of_urine" , 5)
    ds = ds.replace("dischromic__patches" , 6)
    ds = ds.replace("spotting__urination" , 6)        
    return ds
    
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data['Disease']=le.fit_transform(data['Disease'])
```

```python
import numpy as np
x=np.array(data.drop(['Disease'],axis=1))
y=np.reshape(np.array(data['Disease']),(-1,1))
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
disease_classifier_1=RandomForestClassifier(n_estimators=50)
disease_classifier_1.fit(x_train,y_train)
```

```python
disease_classifier_1.score(x_test,y_test)
```
